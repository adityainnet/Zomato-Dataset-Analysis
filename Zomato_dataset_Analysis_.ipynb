```python
df[df['online_order'].astype(str).str.contains("Rated", case=False, na=False)].head()
```

```python
df = df[
    (df['online_order'].isin(['Yes', 'No'])) &
    (df['book_table'].isin(['Yes', 'No']))
]

print("Shape after removing corrupted rows:")
print(df.shape)
```

```python
print(df['online_order'].unique())
print(df['book_table'].unique())
```

```python
print("Before:", df.shape)

df = df.drop_duplicates()

print("After:", df.shape)
```

```python
pd.DataFrame({
    'Missing Values': df.isnull().sum(),
    'Percentage': (df.isnull().sum()/len(df))*100
}).sort_values(by='Percentage', ascending=False)
```

```python
df['rate'].unique()[:20]
```

```python
df['rate'] = df['rate'].astype(str)

df['rate'] = df['rate'].str.replace('/5', '', regex=False)

df['rate'] = df['rate'].replace({
    'NEW': np.nan,
    '-': np.nan,
    'nan': np.nan
})

df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
df['rate'].head()
```

```python
df['rate'].describe()
```

```python
df['votes'].dtype
```

```python
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
```

```python
df['votes'].describe()
```

```python
df['approx_cost(for two people)'].head(20)
```

```python
df['approx_cost(for two people)'] = (
    df['approx_cost(for two people)']
    .astype(str)
    .str.replace(',', '', regex=False)
)

df['approx_cost(for two people)'] = pd.to_numeric(
    df['approx_cost(for two people)'],
    errors='coerce'
)
```

```python
df['approx_cost(for two people)'].describe()
```

```python
df.info()
```

```python
df = df.drop(['phone', 'address'], axis=1)

print(df.shape)
```

```python
df['dish_liked'] = df['dish_liked'].fillna('Not Mentioned')
```

```python
df['location'] = df['location'].fillna(df['location'].mode()[0])

df['rest_type'] = df['rest_type'].fillna(df['rest_type'].mode()[0])

df['cuisines'] = df['cuisines'].fillna(df['cuisines'].mode()[0])

df['listed_in(type)'] = df['listed_in(type)'].fillna(df['listed_in(type)'].mode()[0])
```

```python
pd.DataFrame({
    'Missing Values': df.isnull().sum(),
    'Percentage': (df.isnull().sum()/len(df))*100
}).sort_values(by='Percentage', ascending=False)
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x='rate',
    bins=20,
    kde=True
)

plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')

plt.show()
```

The distribution of restaurant ratings indicates that the majority of restaurants maintain ratings between 3.5 and 4.2, with the highest concentration around 3.8. Very few restaurants have ratings below 2.5, suggesting generally satisfactory customer experiences. Similarly, only a small percentage of restaurants achieve ratings above 4.5, indicating that exceptional performance is relatively rare.

```python
df['main_cuisine'] = df['cuisines'].str.split(',').str[0]
```

```python
top_cuisines = df['main_cuisine'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_cuisines.values,
    y=top_cuisines.index
)

plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')

plt.show()
```

```python
top_cuisines
```

North Indian cuisine is the most dominant category with 8,754 restaurants, followed by South Indian and Cafe categories. The large number of cafe establishments indicates a strong café culture in Bengaluru. Biryani and Chinese cuisines also maintain significant popularity, reflecting diverse customer preferences. Fast Food, Bakery, and Dessert outlets collectively represent a major segment of the food industry.

```python
df['approx_cost(for two people)'].describe()
```

```python
plt.figure(figsize=(12,6))

sns.histplot(
    df['approx_cost(for two people)'],
    bins=40,
    kde=True
)

plt.title('Restaurant Cost Distribution')
plt.xlabel('Approx Cost for Two People')
plt.ylabel('Number of Restaurants')

plt.show()
```

```python
def price_category(cost):
    if cost <= 300:
        return 'Budget'
    elif cost <= 700:
        return 'Moderate'
    elif cost <= 1500:
        return 'Expensive'
    else:
        return 'Luxury'

df['price_category'] = df['approx_cost(for two people)'].apply(price_category)
```

```python
price_counts = df['price_category'].value_counts()

price_counts
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='price_category',
    order=['Budget','Moderate','Expensive','Luxury']
)

plt.title('Restaurant Price Categories')

plt.show()
```

The restaurant cost distribution is positively skewed, indicating that most restaurants operate in the affordable to moderate pricing range. The Moderate category contains the highest number of restaurants (15,871), followed by Budget restaurants (12,017). Luxury restaurants account for only a small portion of the market, suggesting that Bengaluru's restaurant industry primarily caters to price-conscious customers.

```python
corr = df[['rate','votes','approx_cost(for two people)']].corr()

corr
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap')

plt.show()
```

BTM emerged as the most restaurant-dense location with 2,913 restaurants. IT-centric areas such as Whitefield, HSR, Electronic City, and Bellandur also recorded high restaurant counts, indicating strong food demand driven by working professionals. Koramangala and Indiranagar remain major dining destinations due to their diverse restaurant ecosystems.

Correlation analysis revealed a moderate positive relationship between restaurant ratings and votes (0.43), suggesting that highly rated restaurants tend to attract greater customer engagement. Restaurant cost also showed moderate positive correlations with ratings (0.38) and votes (0.38), indicating that premium restaurants generally receive better ratings and more customer reviews.
